```r
data.sorted = lapply(data, function(x) x[order(x$Freq, decreasing=T), ])
```

```r
for(index in 1:length(data)){
    a = data.sorted[[index]]
    a.name = names(data.sorted[index])
    a.name =  cData.get(a.name)$name

    a$name = sapply(a$x, function(x) cData.get(x)$name)
    a$p = sapply(a$Freq, function(x) x/max(a$Freq))

    print(a.name)
    print(head(a[, 3:4],5))
}
```
